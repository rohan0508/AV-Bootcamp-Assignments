```python
### Converting the datatype International plan variable into boolean
for index, plan in enumerate(data['International plan']):
    if plan == 'Yes':
        data['International plan'].iloc[index] = True
    elif plan == 'No':
        data['International plan'].iloc[index] = False

data['International plan'] = data['International plan'].astype(bool)
```

### Q 7: Generate additional feature named F_X such that :-
* International plan is True then the value of F_X is 1
* If International plan is False and Account Length is more than the average Account length then F_X is 1
* If Number vmail messages is greater than 75th percentile and Customer Service calls is greater than 1 then F_X is 1
* In all other cases F_X is 0

```python
# calculating average account length, vmail_messages_75_percentile
average_account_length = data['Account length'].mean()
vmail_messages_75_percentile = data['Number vmail messages'].quantile(0.75)
```

```python
# resetting index of the dataframe
data.reset_index(drop=True)
```

```python
#instantiating F_X column
data['F_X'] = pd.Series(dtype=float)
```

```python
#setting F_X column based on conditions
for index in range(data.shape[0]):
    if data.loc[index, 'International plan']:
        data.loc[index, 'F_X'] = 1
    elif data.loc[index, 'International plan'] == False and data.loc[index, 'Account length'] > average_account_length:
        data.loc[index, 'F_X'] = 1
    elif data.loc[index, 'Number vmail messages'] > vmail_messages_75_percentile and \
                            data.loc[index, 'Customer service calls'] > 1:
        data.loc[index, 'F_X'] = 1
    else:
        data.loc[index, 'F_X'] = 0
        
data.F_X = data.F_X.astype(int)
```

```python
# testing for a condition -- must have only 1 as value
data[(data['Number vmail messages'] > vmail_messages_75_percentile) & (data['Customer service calls'] > 1)]['F_X'].value_counts()
```
